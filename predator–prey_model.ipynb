## 1. Introduction

On souhaite décrire l'évolution de deux populations en interaction: une population de proies
(par exemple des lièvres)
et une population de prédateurs (par exemple des lynx).  On note respectivement $x(t)$ et $y(t)$ les effectifs des
proies et prédateurs 
(normalisés par des valeurs de référence) à l'instant $t$.
On peut modéliser l'évolution de ces populations
à l'aide d'équations différentielles de la forme:
\begin{equation}
\left\{
\begin{array}{lll}
x'(t)&=& x(t) \, f(x(t),y(t),t) \\
y'(t) &=& y(t) \, g(x(t),y(t),t)
\end{array}
\right.
\end{equation}
où $f,g$ sont des fonctions $C^1$ qui décrivent les taux de croissance de chaque population.
La fonction $y\mapsto f(x,y,t)$ est décroissante (terme de prédation) et
$x\mapsto g(x,y,t)$ est croissante.
Par ailleurs, on suppose les fonctions 
$x\mapsto f(x,y,t)$ et $y\mapsto g(x,y,t)$ décroissantes
si on veut prendre en compte une quantité limitée de nourriture.
La dépendance de $f,g$ par rapport au temps permet
de tenir compte des variations saisonnières
de la quantité de nourriture et des taux de natalité.

## 2. Equation de Lotka-Volterra

Il s'agit d'une version très simplifiée du modèle précédent, dans laquelle 
la population de proies dispose de ressources illimitées et 
l'équation différentielle est autonome, i.e. les fonctions $f,g$ ne
dépendent pas de $t$: 
\begin{equation}
\left\{
\begin{array}{lll}
x'(t)&=& x(t) ( 1- y(t) ) \\
y'(t) &=& \alpha\, y(t)(x(t)  - 1).
\end{array}
\right.
\end{equation}
Le paramètre $\alpha$ est strictement positif.
Pour les calculs numériques, on fixera $\alpha =1$.

### 2.1. Analyse du modèle

#### Question 1 
Calculer la solution $(x(t),y(t))$ pour la condition initiale
$(x(0),y(0))=(0,y_0)$ avec $y_0 \geq 0$. Quel est son comportement quand $t \rightarrow +\infty$ ?
Interpréter ce résultat.

#### Réponse 1
On pose : $u = \pmatrix {x \\ y}$, le système devient :
\begin{equation}
\begin{cases}
u'\left(t\right) = f\left(u(t)\right) \\
\\
u\left(0\right) = \pmatrix {0 \\ y_{0}}
\end{cases}
\qquad \text{avec}
\qquad f\pmatrix{x \\ y} = \pmatrix{x - xy \\ \alpha xy - \alpha y}
\\
\end{equation}



La fonction $f$ étant polynomiale en ses coordonnées, elle est donc de classe $C^{1}$ sur $\mathbb{R}^{2}$ et est localement lipschitzienne, il y'a donc au moins existence locale des solutions (par théorême de Cauchy-Lipchitz). Et on a de plus :

Si $x_0 = 0$ alors $\forall t \in \mathbb{R}^{+} : x(t) = 0.$

Montrons alors que $\pmatrix {0 \\ y_0 e^{-\alpha t}}$  est solution de l'équation. On a :

\begin{align*}
\pmatrix {0 \\ y_0 e^{-\alpha t}}'  =& \pmatrix {0 \\ -\alpha y_0 e^{-\alpha t}} \\
\\
 =& \pmatrix {0 \times (1 - y_0 e^{-\alpha t}) \\ \alpha y_0 e^{-\alpha t} \times \left(0 - 1\right)}
\end{align*}

Ainsi, par le théorême de Cauchy-Lipschitz qui assure l'unicité de la solution, l'unique solution de l'équation est :

\begin{equation}
\boxed{
\pmatrix {0 \\ y_0 e^{-\alpha t}}
}
\end{equation}

On a :

\begin{equation}
\boxed{\lim_{t\to\infty} \pmatrix {0 \\ y_0 e^{-\alpha t}} = \pmatrix {0 \\ 0}}
\end{equation}

Ceci traduit l'extinction des prédateurs lorsqu'il n'y a plus de proies, et donc de nourriture. Le modèle parait réaliste en ce sens là.


#### Question 2 
Mêmes questions pour la condition initiale $(x(0),y(0))=(x_0 ,0)$ avec $x_0 \geq 0$. 

#### Réponse 2

Inversons maintenant les rôles et supposons que $x_0 \geq 0$ et $y_0 = 0$, on reprend le même raisonnement pour justifier l'existence et l'unicité de la solution.

Montrons ainsi que $\pmatrix {x_0 e^{t} \\ 0}$ est solution de l'équation. On a :

\begin{align*}
\pmatrix {x_0 e^{t} \\ 0}'  =& \pmatrix {x_0 e^{t} \\ 0} \\
\\
 =& \pmatrix {x_0 e^{t} \times (1 - 0) \\ 0 \times \left(x_0 e^{t} - 1\right)}
\end{align*}

Ainsi, par le théorême de Cauchy-Lipschitz qui assure l'unicité de la solution, l'unique solution de l'équation est :

\begin{equation}
\boxed{
\pmatrix  {x_0 e^{t}\\ 0}
}
\end{equation}

On a :

\begin{equation}
\boxed{\lim_{t\to\infty} \pmatrix {x_0 e^{t} \\ 0} = \pmatrix {+\infty \\ 0}}
\end{equation}

Quand il n'y a plus de prédateurs, les proies se reproduisent à l'infini. Cet effet est dû au fait que la fonction $f$ définie dans l'énoncé, et correspondant au taux de croissance de la population des proies, est prise comme étant constante. Ainsi, la population des proies croît à l'infini. En ce sens là, et vu les hypothèses considérées, cette limite semble raisonnable. Mais elle ne prend pas en compte le fait que plus la population des proies augmente, plus la competitivité sur la nourriture augmente, et donc le taux de croissance décroit.

#### Question 3
Déterminer les solutions d'équilibre, i.e. les solutions $(x,y)$ indépendantes de $t$.

#### Réponse 3

Soit (x, y) $\in \mathbb{R}^{2}$, on a :
\begin{array}{rcl}
\text{(x, y) est un point d'équilibre} & \Leftrightarrow & f(x,y) = 0
\\ & \Leftrightarrow & \begin{cases}
x-xy = 0
\\
\alpha xy - \alpha y = 0
\end{cases}
\\ & \Leftrightarrow & \begin{cases}
x(1-y) = 0
\\
\alpha y(x-1)= 0
\end{cases}
\\ & \Leftrightarrow & \boxed{\pmatrix  {x = 0\\ y = 0} ou \pmatrix  {x = 1\\ y = 1}}
\end{array}

#### Question 4
Une fonction $H\, : \, ]0,+\infty[ \times ]0,+\infty[ \rightarrow \mathbb{R}$
est appelée intégrale première de l'équation différentielle
lorsque pour toute solution $(x(t),y(t))$ on a $\frac{d}{dt} H(x(t),y(t))=0$.
Vérifier que 
$$
H(x,y)=\alpha\, (x-\ln{x})+y-\ln{y}
$$
est une intégrale première.

#### Réponse 4

Justifions d'abord que si $x_0 > t$ alors $\forall t > 0, x(t) > 0$, et de même pour $y(t)$.

En effet, si $x_0 > t$ et qu'il existe t tel que $x(t) < 0$, alors le théorême de valeurs intérmediaires donne l'existence d'un point où $x(t)$ s'annule, or comme nous l'avons montré, ceci impliquerait que $\forall t, x(t) = 0$, ce qui contredit la supposition de départ. Le même raisonnement est utilisé pour $y(t)$.


Soit $(x(t),y(t))$ une solution, on a :
\begin{align*}
\frac{d}{dt} H(x(t),y(t)) &= \frac{d}{dt} \left(\alpha\, (x(t)-\ln{x(t)})+y(t)-\ln{y(t)}\right)
\\ &= \alpha\left(x'(t) - \frac{x'(t)}{x(t)}\right) + y'(t) - \frac{y'(t)}{y(t)}
\\ &= \alpha\left(x-xy - \frac{x-xy}{x}\right) + \alpha xy - \alpha y - \frac{\alpha xy - \alpha y}{y}
\\ &= \alpha x - \alpha xy - \alpha + \alpha y + \alpha xy - \alpha y - \alpha x + \alpha
\\ &= 0
\end{align*}

D'où $H$ est une intégrale première.

#### Question 6
Quand on considère 
des solutions $(x(t),y(t))$
de conditions initiales $(x(0),y(0))$ strictement positives,
que peut-on en déduire pour les trajectoires
$\gamma = \{ (x(t),y(t)),\, t \in \mathbb{R} \} \subset {\left(\mathbb{R}^+ \right)}^2 $
ainsi que pour les fonctions $t \mapsto (x(t),y(t))$ ?

#### Réponse 6

On a déjà montré que si $(x(0),y(0))$ sont strictement positives, alors $\forall t > 0: (x(t),y(t)) > 0$, on a donc bien $\gamma \subset {\left(\mathbb{R}^+ \right)}^2 $.

Intéressons-nous à l'allure de la trajectoire. Nous avons déjà montré l'existence de deux points d'équilibre, on partage donc le plan autour de ces deux points d'équilibres en quatres zones :

\begin{equation}
A_1 = \left\{(x,y) \in \mathbb{R}^{2} , x \leq 1\text{ et } y \leq 1  \right\}
\\ A_2 = \left\{(x,y) \in \mathbb{R}^{2} , x > 1\text{ et } y \leq 1  \right\}
\\ A_3 = \left\{(x,y) \in \mathbb{R}^{2} , x > 1\text{ et } y > 1  \right\}
\\ A_4 = \left\{(x,y) \in \mathbb{R}^{2} , x \leq 1\text{ et } y > 1 \right\}
\end{equation}

Supposons que le mouvement commence dans la zone $A_1$ (il suffira de permuter les ensembles sinon), on a donc dans cette zone:

\begin{align*}
x \leq 1\text{ et } y \leq 1 &\implies xy \leq x\text{ et }xy \leq y
\\ &\implies x' \geq 0\text{ et }y' \leq 0
\\ &\implies \text{x va en croissant tandis que y va en décroissant}
\end{align*}

Supposons que le mouvement reste dans cette zone. Puisque les valeurs de x et y sont toutes les deux bornées dans celle-ci, $(x(t),y(t))$ converge. Sa limite est donc nécéssairement un point d'équilibre. Mais $x(t)$ est croissante, donc devrait converger vers une borne supérieure, tandis que $y(t)$ est décroissant, donc devrait converger vers une borne inférieure. Or les seuls points d'équilibre sont (1,1) et (0,0). Il y'a absurdité. Ainsi : le mouvement quitte nécéssairement la zone $A_1$ à partir d'un instant $t_1$ avec $x(t_1) = 1$ et $y(t_1) < 1$. Et puisque x est $C^{1}$, on a alors sur un voisinage de droite $x'(t_1) > 0$ et on rentre alors dans la zone $A_2$.

On démontre de la même manière qu'on part de $A_2$ vers $A_3$, or cette fois-ci, x n'est plus majorée. Montrons qu'elle est tout de même bornée.

Soit la fonction : $\phi (x) = x - \ln x$ On a :
\begin{align*}
& \frac{d \phi}{dx} \left(x\right) = 1 - \frac{1}{x} \leq 0\text{ sur [0,1] et > 0 sur ]1,}+\infty]
\\ & \frac{d^2 \phi}{dx^2} \left(x\right) = \frac{1}{x^2} > 0
\end{align*}

La fonction $\phi$ est donc convexe, et elle est minorée puisque décroissante puis croissante. On remarque que $$ H(x) = \alpha \phi(x) + \phi(y)$$

On sait aussi par la question 4 que H est une intégrale première du mouvement. Supposons qu'il existe k et un intervalle de temps J tel que : $\forall t \in J, H(x(t), y(t)) = k$. L'existence est assurée par le théorême de Cauchy-Lipschitz, on sait que $\phi(x(t)) \geq 1$ donc 

$$\phi(y(t)) = H(x,y) - \alpha \phi(x) \leq k - \alpha$$. 

Mais nous avons :

$$\lim_{t\to0} \phi(y(t)) = \lim_{t\to\infty} \phi(y(t)) = +\infty$$

Et donc, y doit nécessairement être borné. On utilise le même raisonnement pour montrer que x est aussi bornée.

Finalement, puisque x et y sont donc bien bornées dans $A_2$, il suffit de réappliquer le raisonnement fait sur $A_1$. Nous montrons finalement le résultat suivant :

$$\boxed{\text{Le mouvement se fait de la zone 1 vers la 2, de la zone 2 vers la 3, de la zone 3 vers la 4, et de la zone 4 vers la 1.}}$$

Montrons maintenant que le mouvement est périodique. On remarque que la fonction $y \rightarrow H(1,y) = \alpha + y - ln(y)$ est strictement croissante sur l'intervalle $]1, +\infty[$. Mais nous avons déjà montré que notre trajectoire tourne, il y'a donc une infinité d'intersections avec la demi-droite x = 1, y > 1. Et puisque H est constant le long de la trajectoire, on revient au même point de cette demi-droite.

On note alors $t_2$ et $t_1$ deux instants quelconques durant lesquels la trajectoire coupe la demi-droite. Et on remarque que le couple $\left(x\left(t+t_2-t_1\right),y\left(t+t_2-t_1)\right )\right)$ est aussi une solution de notre système, et qu'il coincide avec (x,y) en $t_1$. Par unicité donnée par le théorême de Cauchy-Lipschitz :

$$\boxed{\forall t \in \mathbb{R} \pmatrix {x(t+t_2-t_1) \\ y(t+t_2-t_1)} = \pmatrix {x(t) \\ y(t)}}$$

## Conclusion :

Le mouvement est périodique et se fait de $A_1$ vers $A_2$ vers $A_3$ vers $A_4$.

### 2.2. Intégration Numérique

#### 2.2.1. Schéma d'Euler explicite
Afin d'approcher numériquement les solutions du modèle de Lotka-Volterra
pour $t \in [0,t_{\mathrm{max}}]$, 
on discrétise l'équation à l'aide du schéma d'Euler explicite:
\begin{equation}
\left\{
\begin{array}{lll}
x_{k+1}&=& x_k + h\, x_k ( 1- y_k ) \\
y_{k+1} &=& y_k + h\, y_k (x_k  - 1)
\end{array}
\right.
\end{equation}
où $h >0$ est le pas d'intégration en temps et
$(x_k,y_k)$ une approximation de $(x(t_k),y (t_k))$ en $t_k = k\, h$.
On fixe $h=t_{\mathrm{max}}/N$ où $N$ est un entier $\geq 1$.

#### Question 7
Pour $t_{\mathrm{max}}=120$, $(x(0),y(0))=(2,1)$ 
et différentes valeurs du pas de temps $h$, 
tracer les trajectoires 
$(x_k,y_k)_{0 \leq k \leq N} \subset {\left(\mathbb{R}^+ \right)}^2$
ainsi que les graphes de $x_k$, $y_k$ en fonction de $t_k$. 
Qu'observez-vous ? Comparer ces résultats à l'étude théorique 
de la question 6.

```python
import numpy as np
def euler_explicite(tmax, h, x0, y0):
    X = [x0]
    Y = [y0]
    T = np.arange(0, tmax, h)
    for i in T[1:]:
        X.append(X[-1]*(1+h*(1-Y[-1])))
        Y.append(Y[-1]*(1+h*(X[-2]-1)))
    fig, axs = plt.subplots(1, 3, figsize=(20, 5))
    fig.suptitle(f"Courbes pour h={h}", size =30, y=1.05)
    axs[0].plot(T, X)
    axs[0].set_xlabel('temps tk')
    axs[0].set_ylabel('effectifs proies xk')
    axs[0].set_title("xk en fonction de tk", size = 20)
    axs[1].plot(T, Y)
    axs[1].set_xlabel('temps tk')
    axs[1].set_ylabel('effectifs prédateurs yk')
    axs[1].set_title("yk en fonction de tk", size = 20)
    #plt.plot(T,X)
    #plt.plot(T,Y)
    axs[2].plot(X,Y)
    axs[2].set_xlabel('effectifs proies xk')
    axs[2].set_ylabel('effectifs prédateurs yk')
    axs[2].set_title("yk en fonction de xk", size = 20)
    plt.show()
euler_explicite(120,0.0001,2,1)
euler_explicite(120,0.001,2,1)
euler_explicite(120,0.01,2,1)
euler_explicite(120,0.03,2,1)
euler_explicite(120,0.06,2,1)
```

On peut voir que plus le pas de discrétisation h est petit, plus le schémas semblent être périodique et l'erreur est se réduire. Le schéma est donc consistent.

Cependant plus h est grand et plus le schéma devient instable, l'erreur semble doubler à chaque fois que h est multiplié par 2. (ymax passe de 10 à 20 quand h passe de 0.03 à 0.06).

L'erreur semble doubler à chaque période. Donc la méthode est instable.

PLus h est grand plus le portrait de phase se rapproche d'un triangle rectagle dont les cotés sont parallèle au axes x et y.

Quand h est petit on reconnait la forme des courbes de niveau de la question 5 et les trajectoires semblent périodiques, ce qui va dans le sens du résultat de la question 6.

Reprendre la question 7 pour le schéma d'Euler implicite.

```python
import math as math
def euler_implicite(tmax, h, x0, y0):
    X = [x0]
    Y = [y0]
    T = np.arange(0, tmax, h)
    for i in T[1:]:
        delta = (1-h**2-h*(X[-1]+Y[-1]))**2 + 4*(h**2+h)*Y[-1]*(1-h)
        Y.append(((h**2+h*(X[-1]+Y[-1])-1)+math.sqrt(delta))/(2*(h**2+h)))
        X.append(X[-1]/(1+h*(Y[-1]-1)))
    fig, axs = plt.subplots(1, 3, figsize=(20, 5))
    fig.suptitle(f"Courbes pour h={h}", size =30, y=1.05)
    axs[0].plot(T, X)
    axs[0].set_xlabel('temps tk')
    axs[0].set_ylabel('effectifs proies xk')
    axs[0].set_title("xk en fonction de tk", size = 20)
    axs[1].plot(T, Y)
    axs[1].set_xlabel('temps tk')
    axs[1].set_ylabel('effectifs prédateurs yk')
    axs[1].set_title("yk en fonction de tk", size = 20)
    #plt.plot(T,X)
    #plt.plot(T,Y)
    axs[2].plot(X,Y)
    axs[2].set_xlabel('effectifs proies xk')
    axs[2].set_ylabel('effectifs prédateurs yk')
    axs[2].set_title("yk en fonction de xk", size = 20)
    plt.show()
euler_implicite(120,0.0001,2,1)
euler_implicite(120,0.001,2,1)
euler_implicite(120,0.01,2,1)
euler_implicite(120,0.1,2,1)
euler_implicite(120,1,2,1)
```

#### 2.2.3. Schéma d'Euler semi-implicite
On considère un schéma d'Euler semi-implicite
\begin{equation}
\left\{
\begin{array}{lll}
x_{k+1}-x_k&=&  h\, x_{k} ( 1- y_{k+1} ) \\
y_{k+1} -y_k&=&  h\, y_{k+1} (x_{k}  - 1)
\end{array}
\right.
\end{equation}
dans lequel la variable $x$ est traitée de manière explicite
(évaluée en $t=t_{k}$ dans le second membre de l'équation différentielle)
et $y$ de manière implicite (évaluée en $t=t_{k+1}$).

On peut voir que comme pour le schémas précédent si on fait tendre h vers 0, l'erreur diminue donc le schéma est consistant.

Contrairement au schéma euler-explicite qui diverge, ici l'erreur est borné par la distance entre (x0, y0) et le point fixe (1,1) $\sqrt{(x0-1)^2+(y0-1)^2}$. Le schéma est donc stable.

D'après le théorème fondamentale, il est donc convergent. Cependant il converge vers le point fixe (1,1) et pas vers la trajectoire périodique.

Pour h petit on peut approximer le portrait de phase à une trajectoire périodique qui valide la question 6.

## 3. Modèle proies-prédateurs avec croissance limitée
On considère l'équation différentielle:
\begin{equation}
\left\{
\begin{array}{lll}
x'(t)&=& x(t) ( 1- y(t) - c(t)\, x(t)) \\
y'(t) &=&  y(t)(x(t)  - 1)
\end{array}
\right.
\end{equation}
où le coefficient $c(t)>0$ 
tient compte du fait que la population de proies est limitée par la quantité de nourriture.
On suppose que la fonction $c$ est $T$-périodique
(variations saisonnières des ressources) avec $T=12$ (l'unité de temps représente un mois).
Dans ce TP on fixera 
$$
c(t)=\lambda\, \left(1+\frac{1}{2}\, \cos{\left( \frac{\pi}{6}\, t \right)}\right),
$$
où $\lambda >0$ est un paramètre.

On discrétise l'équation différentielle par le
schéma d'Euler semi-implicite
\begin{equation}
\left\{
\begin{array}{lll}
x_{k+1}-x_k&=&  h\, x_{k} ( 1- y_{k+1} - c(t_k)\, x_k) \\
y_{k+1} -y_k&=&  h\, y_{k+1} (x_{k}  - 1)
\end{array}
\right.
\end{equation}

#### Question 12
Exprimer  $(x_{k+1},y_{k+1})$ en fonction de $(t_k,x_k,y_{k},h)$.

On a 
$$\begin{array}{rcl}
y_{k+1} -y_k=  h\, y_{k+1} (x_{k}  - 1) & \Leftrightarrow & y_{k+1}(1+h-hx_k) = y_k\\
& \Leftrightarrow & \boxed{y_{k+1} =\dfrac{y_k}{1+h-hx_k}}\\
x_{k+1}-x_k=  h\, x_{k} ( 1- y_{k+1} - c(t_k)\, x_k)  & \Leftrightarrow & x_{k+1}-x_k=  h\, x_{k} \left( 1- \dfrac{y_k}{1+h-hx_k} -c(t_k)x_k\right) \\
& \Leftrightarrow & \boxed{x_{k+1}=  x_{k} \left( 1 + h- hc(t_k)x_k-\dfrac{hy_k}{1+h-hx_k} \right)}\\
& \Leftrightarrow & x_{k+1}=  x_{k} \left[ 1 + h- h\lambda\, \left(1+\frac{1}{2}\, \cos{\left( \frac{\pi}{6}\, t \right)}\right)x_k-\dfrac{hy_k}{1+h-hx_k}  \right]
\end{array}
$$

#### Question 13
On fixe $t_{\mathrm{max}}=120$ et $h=10^{-3}$.
Pour différentes valeurs du paramètre $\lambda$,
tracer les graphes de $x_k$, $y_k$ en fonction de $t_k$
pour différentes conditions initiales $(x(0),y(0))$. 
Qu'observez-vous ? 
Interpréter le comportement des solutions lorsque
$t\rightarrow +\infty$
en terme d'évolution des populations, ainsi que l'effet du paramètre $\lambda$.

```python
def c(t, l):
    return l*(1+0.5*math.cos(math.pi/6*t))

def euler_para(tmax, h, x0, y0, l):
    X = [x0]
    Y = [y0]
    T = np.arange(0, tmax, h)
    for t in T[1:]:
        Y.append(Y[-1]/(1+h-h*X[-1]))
        X.append(X[-1]*(1+h-h*c(t, l)*X[-1]-h*Y[-1]))
    fig, axs = plt.subplots(1, 3, figsize=(20, 5))
    fig.suptitle(f"Courbes pour x0={x0}, y0={y0}, λ={l}", size =30, y=1.05)
    axs[0].plot(T, X)
    axs[0].set_xlabel('temps tk')
    axs[0].set_ylabel('effectifs proies xk')
    axs[0].set_title("xk en fonction de tk", size = 20)
    axs[1].plot(T, Y)
    axs[1].set_xlabel('temps tk')
    axs[1].set_ylabel('effectifs prédateurs yk')
    axs[1].set_title("yk en fonction de tk", size = 20)
    #plt.plot(T,X)
    #plt.plot(T,Y)
    axs[2].plot(X,Y)
    axs[2].set_xlabel('effectifs proies xk')
    axs[2].set_ylabel('effectifs prédateurs yk')
    axs[2].set_title("yk en fonction de xk", size = 20)
    plt.show()
```

### Influence de lambda

```python
euler_para(120,0.001,2,1,0.00001)
euler_para(120,0.001,2,1,0.001)
euler_para(120,0.001,2,1,0.1)
euler_para(120,0.001,2,1,0.2)
euler_para(120,0.001,2,1,2)
```

On peut voir que le schéma converge au bout d'un moment vers une trajectoire ellipsoïdal centrer sur (1,yc).

Plus $\lambda$ est petit plus il converge lentement vers cette trajectoire.

Plus $\lambda$ est grand plus yc est petit jusqu'à à avoir une extinction des prédateurs à partir d'un certain $\lambda$

## Conclusion

 Le modèle de Lotka-Volterra est fréquemment utilisé pour décrire l'évolution des systèmes écologiques où deux éspèces interagissent. Le modèle que nous avons adopté est un modèle simplifié et se construit sur certaines suppositions : (1) Seulement deux éspèces existent; (2) Les proies naissent et meurent ensuite soit par prédation soit par manque de nourriture. (3) Les prédateurs naissent et leur taux de naissance est positivement affecté par leur taux de prédation, et meurent naturellement.
 
 Les caractéristiques de ce modèle sont que les deux populations s'influencent mutuellement, d'où la modélisation sous forme de système. La taille de la population des proies a un effet positif sur la taille de la population des prédateurs, et la taille de la population des prédateurs a un effet négatif sur la taille de la population des proies.
 
 Les résultats produits par le modèle sont satisfaisants. Ils confirment certains résultats instinctifs (Extinction des prédateurs si manque de proies etc.). Les populations des proies et des prédateurs se maintiennent effectivement les unes les autres, ce qui est démontré par le fait que la trajectoire est périodique et qu'elle forme, à un homéomorphisme près, un cercle centré en (1,1).
 
 Reste alors à généraliser le modèle, en introduisant par exemple un facteur représentant l'influence de l'Homme sur les populations, ou bien en introduisant de nouvelles éspèces : des proies pour les proies ou des prédateurs pour les prédateurs, jusqu'à créer une chaine alimentaire.
